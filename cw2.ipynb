```python
import pandas as pd
#Reading files and merging
df1 = pd.read_csv('FRATNPdeath.txt', header=0)
df2 = pd.read_csv('USAdeath.txt', header=0)
df3 = pd.read_csv('ITAdeath.txt', header=0)
df4 = pd.read_csv('DEUTNPdeath.txt', header=0)
df5 = pd.read_csv('ESPdeath.txt', header=0)
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
#Dropping useless coloumns
df.drop(['RefCode', 'Access', 'NoteCode1', 'NoteCode2', 'NoteCode3', 'LDB', 'YearReg', 'Lexis',  'Area', 'YearInterval', 'AgeInterval'], axis=1, inplace=True)

print(df)
```

```python
#Dropping empty ages that arent numbers and only keeping age 0-21 young to young adults
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df = df.dropna(subset=['Age'])
df = df.loc[(df['Age'] >= 0) & (df['Age'] <= 21)]
df
```

```python
#only keeping the begining of H1N1 influenza and Covid-19
df = df[(df['Year'].between(2009, 2009)) | (df['Year'].between(2019, 2019))]
df
```

```python
#Renaming Country appropriately 
df.loc[df['PopName'] == 'FRATNP', 'PopName'] = 'FRA'
df.loc[df['PopName'] == 'DEUTNP', 'PopName'] = 'GER'
df.loc[df['PopName'] == 'ESPTNP', 'PopName'] = 'ESP'
```

```python
#Saving as Excel to use in Tableau
df.to_excel('Newdata.xlsx', index = False)
```
